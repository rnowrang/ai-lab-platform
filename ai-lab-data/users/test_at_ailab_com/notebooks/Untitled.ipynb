```python
import mlflow
import mlflow.pytorch
import torch
import torch.nn as nn
import torch.optim as optim

# 1. Set the MLflow tracking URI to point to the central server
#    (This is already configured in your environment)
mlflow.set_tracking_uri("http://localhost:5000")

# 2. Start an MLflow run
with mlflow.start_run(run_name="My First PyTorch Experiment"):
    
    # 3. Log hyperparameters
    learning_rate = 0.01
    epochs = 10
    mlflow.log_param("learning_rate", learning_rate)
    mlflow.log_param("epochs", epochs)
    
    # Simple model and data
    model = nn.Linear(10, 1)
    optimizer = optim.SGD(model.parameters(), lr=learning_rate)
    X = torch.randn(100, 10)
    y = torch.randn(100, 1)
    
    # Training loop
    for epoch in range(epochs):
        optimizer.zero_grad()
        outputs = model(X)
        loss = nn.functional.mse_loss(outputs, y)
        loss.backward()
        optimizer.step()
        
        # 4. Log metrics
        mlflow.log_metric("loss", loss.item(), step=epoch)

    # 5. Log the trained model
    mlflow.pytorch.log_model(model, "model")

    print("âœ… Experiment logged to MLflow!")
```
